```python
effect_df = pd.read_csv('/Users/ajaobasit/Downloads/cashew_effects.csv',
                         sep='|',
                         skipinitialspace=True)
effect_df = effects_df.dropna()

constraints_df =pd.read_csv('/Users/ajaobasit/Downloads/cashew_Constraints.csv',
                            sep='|',
                         skipinitialspace=True)

constraints_df = constraints_df.dropna()

print("Effects Data:")
print(effects_df)

print("\nConstraints Data:")
print(constraints_df)

    
```

# 

```python
constraints_df.columns = ['Constraint', 'Percentage']
effects_df.columns = ['Effect_Type', 'Statement', 'Weighted_Mean_Score', 'Rank']


plt.figure(figsize=(10, 6))
sns.barplot(data=constraints_df, 
            x='Percentage', 
            y='Constraint',
            color='steelblue')
plt.title('Constraints to Cashew Nut Marketing in Ogbomoso')
plt.xlabel('Percentage of Respondents (%)')
plt.ylabel('Constraint')
plt.tight_layout()
plt.show()
```

```python
negative_df = effects_df[effects_df['Effect_Type'] == 'Negative']

plt.figure(figsize=(10, 6))
sns.barplot(data=negative_df,
            x='Weighted_Mean_Score',
            y='Statement',
            color='crimson')
plt.title('Negative Effects of Cash Scarcity on Cashew Nut Marketing')
plt.xlabel('Weighted Mean Score')
plt.ylabel('Effect Statement')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

```python
positive_df = effects_df[effects_df['Effect_Type'] == 'Positive']

plt.figure(figsize=(10, 6))
sns.barplot(data=positive_df,
            x='Weighted_Mean_Score',
            y='Statement',
            color='green')
plt.title('Positive Effects of Cash Scarcity on Cashew Nut Marketing')
plt.xlabel('weighted Mean Score')
plt.ylabel('Effect Statement')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

```python
neutral_df = effects_df[effects_df['Effect_Type'] == 'Neutral']

plt.figure(figsize=(10, 6))
sns.barplot(data=neutral_df,
            x='Weighted_Mean_Score',
            y='Statement',
            color='gray')
plt.title('Neutral Effects of Cash Scarcity on Cashew Nut Marketing')
plt.xlabel('Weighted Mean Score')
plt.ylabel('Effect statement')
plt.xlim(0, 5)
plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(12, 8))
colors = {'Negative': 'crimson',
          'Positive': 'green',
          'Neutral': 'grey'}

for effect_type, color in colors.items():
    subset = effects_df[effects_df['Effect_Type'] == effect_type]
    plt.barh(subset['Statement'],
             subset['Weighted_Mean_Score'],
             color=color,
             label=effect_type)

plt.title('All Effects of Cash Scarcity on Cashew Nut Marketing')
plt.xlabel('Weighted Mean Score')
plt.ylabel('Statement')
plt.xlim(0, 5)
plt.legend()
plt.tight_layout()
plt.show()
    
```

## Analysis Summary

###  Key Finding 1 - Marketing Constraints
Poor market information (93.6%) was the biggest constraint facing cashew nut traders in Ogbomoso. Traders lack access to market data needed to make informed business decisions.

###  Key Findings 2 - Negative Effects
Reduced sales volume (WMS:3.05) was the most severe negative effect of the 2023 cash scarcity policy, showing direct impact on traders income and livelihoods.

###  Key Findings 3 - Positive Adaptation
Despite hardships, traders adapted by diversifying payment option (WMS 3.27), Showing resilience exists in rural markets even under policy pressure.

## Recommendations

### For Policymakers
Future cash-related policies should include emergency 
cash distribution mechanisms for rural markets to protect 
traders from income loss during implementation.

### For NGOs and Development Organizations
Invest in market information systems for rural traders. 
93.6% cited poor market information as their biggest 
constraint — this is an addressable gap.

### For Extension Officers
Train rural traders on digital payment platforms to 
build on the adaptation already happening. 3.27 WMS 
shows traders are willing to diversify payment methods.

## Conclusion
The 2023 cash scarcity policy disproportionately affected 
rural cashew nut traders in Ogbomoso. While traders showed 
resilience by adapting to digital payments, structural 
constraints like poor market information remain unaddressed. 
Evidence-based rural development interventions are needed.
